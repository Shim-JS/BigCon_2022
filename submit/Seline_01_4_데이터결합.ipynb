```python
result = pd.merge(df1,df2,how = "outer", on = col,indicator=True)
result = result.rename(columns={"count_cust_x" : "count_cust_resident","count_cust_y" : "count_cust_activity"})
result2 = result.drop(["count_cust_activity","_merge"],axis=1)
result2
```

```python
result3 = pd.merge(result2,df3,how = "outer", on = col,indicator=False)
result3 = result3.rename(columns={"count_cust" : "count_cust_ra"})
result3.loc[:,'count_cust_activity'] = result.loc[:,'count_cust_activity']
result3
```
