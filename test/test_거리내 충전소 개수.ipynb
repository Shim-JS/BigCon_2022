```python
# 직선거리 구하기위한 패키지 haversine 설치
!pip install haversine 

from haversine import haversine
```

```python
import pandas as pd

df = pd.read_csv("../../data/extra/1_prepro_전기차충전소.csv",encoding = "cp949")
df
```

```python
df2 = pd.read_csv("../../data/merge_data_3.csv",encoding = "cp949")
mohyun = df2.loc[df2.adng_nm == "모현읍"]
mohyun = mohyun.reset_index(drop = True)
mohyun
```

```python
# 직선거리 정의
for i in range(len(mohyun)):
  count_temp = 0
  # print(i)
  for t in range(len(df)):
    chung = (df.loc[t,"lat"],df.loc[t,"lng"])
    mohyun_temp = (mohyun.loc[i,"cell_ycrd"],mohyun.loc[i,"cell_xcrd"])
    temp = haversine(chung,mohyun_temp)
    # 직선거리 500m 이하인 지점 카운트
    if temp <= 0.5:
      count_temp += 1

  mohyun.loc[i,"500m이내_전기차충전소"] = count_temp


display(mohyun)
```
